#### –¢—Ä–∏–≤–∏–∞–ª—å–Ω–æ–µ —Ä–∞–≤–Ω–æ–≤–µ—Å–∏–µ ùëÄ0(0, 0)

```julia
using Plots
using DifferentialEquations
using ParameterizedFunctions
using ImplicitEquations
```

–ó–∞–¥–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è (–∏–∑ —Ç–µ–∫—Å—Ç–æ–≤–∫–∏–º)

```julia
g = 1
e = 0.01
a = 0.87
d = 0.002
x0 = 0
y0 = 0
```

–ó–∞–¥–∞–µ–º —Å–∏—Å—Ç–µ–º—É –¥–∏—Ñ. —É—Ä–∞–≤–Ω–µ–Ω–∏–π

```julia
pp! = @ode_def PP begin
    dx = x - ((x*y)/(1+a*x)) - e*x^2
    dy = -g*y+((x*y)/(1+a*x))-d*y^2
end g e a d
```

```julia
M = [x0,y0]
```

```julia
param=[g, e, a, d]
```

```julia
timespan = (0.0,100.0)
```

```julia
problem = ODEProblem(pp!, M, timespan, param)
```

```julia
solution1 = solve(problem)
```

```julia
plot(solution1, title = "–î–µ—Ç–µ—Ä–º–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –º–æ–¥–µ–ª—å –•–∏—â–Ω–∏–∫-–ñ–µ—Ä—Ç–≤–∞", xlabel = "t", ylabel = "x, y", label=["–ñ–µ—Ä—Ç–≤–∞ (x)" "–•–∏—â–Ω–∏–∫ (y)"])
```

```julia
plot(solution1, vars=(1,2), xaxis="–ñ–µ—Ä—Ç–≤–∞", yaxis="–•–∏—â–Ω–∏–∫",legend=false)
```
